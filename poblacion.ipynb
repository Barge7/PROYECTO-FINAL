```python
url = 'https://www.mundodeportivo.com/uncomo/educacion/articulo/paises-y-capitales-de-europa-25700.html'
```

```python
driver=webdriver.Chrome(PATH)
driver.get(url)
```

```python
driver.find_element(By.XPATH, '//*[@id="didomi-notice-agree-button"]').click()
```

```python
todos = driver.find_elements(By.TAG_NAME, 'li')
```

```python
lista= []
for e in todos[11:50]:
    lista.append(e.text)
```

```python
dictio = {'poblacion': lista}
```

```python
data = pd.DataFrame(dictio)
```

```python
data = pd.read_excel('poblacion.xlsx')
```

```python
pobla =  []
ciudad = []
for e in data.poblacion:
    e = e.replace('.', '')
    pobla.append(int(re.findall('[0-9+]+.', e)[0]))
    e = e.lower()
    e = e.split('(')
    e = e[1].split(')')
    ciudad.append(e[0])
    
```

```python
data['ciudad']= ciudad
data['poblacion']= pobla
```
